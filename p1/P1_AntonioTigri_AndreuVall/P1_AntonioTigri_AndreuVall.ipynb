### 1.4 Images manipulation

a) Create an RGB black image, 10% larger than the Lenna image. Afterwards, embed the Lenna image in the center. Show the framed image.

**Note:** Be carefull with the number of channels.

**Note:** Print the original image shape and the new image size to check if the sizes of the second one are 10% larger then the first one. 


```python
'''
First we create a new tuple with the shape of the black image (frame) where we just 
multiply the x and y axis of the lenna image shape by 1.1 thus getting a size 10% larger.
The z shape is the same (3 channels for RGB)
'''
increased_shape = (np.array(lenna.shape) * [1.1, 1.1, 1]).round().astype('int')

# Then we create a new black image with the defined shape where we will add the lenna image at the center
framed_lenna = np.zeros(increased_shape, dtype='uint8')

'''
In order for lenna to be at the center, we will define a start index and an end index for the
framed_lenna image where we will put lenna. The idea is to divide the 10% in each axis into two
pices both at the start and at the end. To do this we just get the difference between the frame
image and lenna image and divide it by two (since the operation works for both axis we can operate
the arrays directly) to get the start index and to get the end axis we just have to add to this start
index the size of the lenna image.
'''
start = (increased_shape - lenna.shape) // 2
end = start + lenna.shape

# Then we display the asked values, we also calculate the ratio and round it to make it easier to read.
print('lenna shape:', lenna.shape)
print('Image Frame Shape:', framed_lenna.shape)
shape_ratio = ((np.array(framed_lenna.shape) / np.array(lenna.shape)) - 1).round(4) * 100
print('X Ratio:', shape_ratio[0], '%')
print('Y Ratio:', shape_ratio[1], '%')

'''
Then we change all the black pixels starting at "start" and ending at "end" for the lenna pixels
as we mentioned in the previous comment block
'''
framed_lenna[start[0]:end[0], start[1]:end[1], start[2]:end[2]] = lenna

# And finally, we show the image like always
plt.title('Framed Lenna')
plt.imshow(framed_lenna)
plt.axis('off')
plt.show()
```

b) Change the frame of the new image to green. Visualize the three images: original, black frame and green frame, in one single figure, with their respective titles.

```python
'''
To change the color we will divide the intervals where the image is black into 4, basically the 
4 exterior bands. Luckily, we can reuse our "start" and "end" indexes for this. The green channel is the 2nd
'''
framed_lenna_green = framed_lenna.copy()
framed_lenna_green[:start[0], :, 1] = 255
framed_lenna_green[:, :start[1], 1] = 255
framed_lenna_green[end[0]:, :, 1] = 255
framed_lenna_green[:, end[1]:, 1] = 255

# Then we create a figure where the images will be displayed
fig, axs = plt.subplots(ncols=3, figsize=(16, 8))

# And then we display all the images without axis
axs[0].title.set_text('Lenna')
axs[0].imshow(lenna)
axs[0].axis('off')

axs[1].title.set_text('Black Framed Lenna')
axs[1].imshow(framed_lenna)
axs[1].axis('off')

axs[2].title.set_text('Green Framed Lenna')
axs[2].imshow(framed_lenna_green)
axs[2].axis('off')

plt.show()
```

d) Apply contrast enhancement to the original image. There are two ways for it: (1) by direct manipulation, and (2) through scikit-image command [rescale_intensity](https://scikit-image.org/docs/stable/api/skimage.exposure.html#skimage.exposure.rescale_intensity). Check the minimal and maximal values of the enhanced image, and compare them with the previous one. Show the two images  without graylevel enhancement.

```python
from skimage.exposure import rescale_intensity

# Save miniumum in a variable in order to compute it only once
minimum, maximum = car_low_contrast.min(), car_low_contrast.max()

# We apply the linear rescaling of [minium, maximum] -> [0, 255] and round it back to uint8.
# It's important to note that if a uint8 matrix is multiplied by 255 then there's an overflow error,
# so we need to multiply after casting the matrix to float (it's done automatically when we divide the matrix).
# Also note that the direct enhanced image is slighly different than the scikit enhanced one,
# because they don't round the results before converting them to uint8 again,
# and we consider that in this case it's better to round
enhanced_direct_im = ((car_low_contrast - minimum) / (maximum - minimum) * 255).round().astype('uint8')
enhanced_scikit_im = rescale_intensity(car_low_contrast)

# We will then display the maximum and minimum of each of the 3 images

print('Maximum cell value:', car_low_contrast.max())
print('Minimum cell value:', car_low_contrast.min(), '\n')

print('Maximum cell value:', enhanced_direct_im.max())
print('Minimum cell value:', enhanced_direct_im.min(), '\n')

print('Maximum cell value:', enhanced_scikit_im.max())
print('Minimum cell value:', enhanced_direct_im.min())

# We create the figure to display both enhancements and then we display them
fig, axs = plt.subplots(ncols=2, figsize=(16, 8))

axs[0].title.set_text('Directly Enhanced Car')
axs[0].imshow(enhanced_direct_im, cmap='gray')
axs[0].axis('off')
axs[1].title.set_text('Scikit Enhanced Car')

axs[1].imshow(enhanced_scikit_im, cmap='gray')
axs[1].axis('off')

plt.show()
```

### 1.6 Masks and logical image manipulation

a) Read the image `images_notebook/circles.bmp` (see figure top) and define 3 images in such a way that each one contains only 1 ball (see figure bottom). Visualize the 3 images in the same figure, each one with its corresponding title.

<img src="images_notebook/ex16.png">

**Note:** Be careful when applying `imshow` since it automatically increases the contrast of the image (putting minimum values to 0 and maximal to 1).

**Note:** Be careful when applying type conversion. Check the ranges of the image values.

```python
import collections

# First, we read the image
circles = io.imread('images_notebook/circles.bmp')

# Since all circles have a solid color, we will make a Counter to extract all the colors of each circle
circle_values = list(collections.Counter(circles.reshape(circles.size)).keys())

# Then we sort them and get rid of the last element because we don't want to have the white color in the array
circle_values.sort()
circle_values = circle_values[:-1]

# Then we create the 3 column figure that will contain the 3 different images
fig, axs = plt.subplots(ncols=len(circle_values), figsize=(16, 8))

# We will use a loop to display them and do the necessary mask computations, as we will loop through the 3 colors
for ax, value, position in zip(axs, circle_values, ['Left', 'Central', 'Right']):
    # We will create a mask in which we will have 1's where the pixels have the same color as the one of the current iteration
    mask = (circles == value)
    # Then we apply the mask by multiplying, and also adding 255 times the opposite of the mask, so that everything that is not
    # the desired ball becomes white.
    circles_tmp = circles * mask + 255 * ~mask
    # and at last, we show the image.
    ax.imshow(circles_tmp.astype('uint8'))
    ax.title.set_text(position + ' circle')
    ax.axis('off')
plt.show()
```

b) Given the images `images_notebook/coat.png` and `images_notebook/model.png` (first two images on the figure below), implement the necessary code to construct the image of the model with the coat (third image). Next find in internet an image with a texture and apply it to obtain an image of the jacket with the found texture (fourth image is only an example).

<img src="images_notebook/ex17d.png">

**Note:** when applying [resize, rotate, etc...](https://scikit-image.org/docs/stable/auto_examples/transform/plot_rescale.html) keep track how the image type and values range change. When necessary, apply type conversion using [img_as_float, img_as_uint, img_as_ubyte.](https://scikit-image.org/docs/stable/api/skimage.html)

**Note: DO NOT USE LOOPS to overlap the images!! It will be penalized.**


```python
coat = io.imread('images_notebook/coat.png')
model = io.imread('images_notebook/model.png')

'''
We need to define a way to tell which pixels are from the coat and which ones are not.
The simplest way to do it would be that 100% black pixels are outside the coat and the others pixels are the coat.
But there are sompe pixels around the edge of the coat that look almost black but have RGB of (2, 2, 1), and we
want to consider them outside of the coat (because they look black to us).

One natural way to tackle this problem is to define a distance between colors. Then we'll compute the distance
from every pixel to pure black and by analyzing the results we'll define a threshold distance to tell apart
the coat, i.e. only pixels with distance(pixel, black) > threshold_distance will be considered from the coat.

We considered first as distance the euclidean distance between tuples (R, G, B), but we later found out that the
euclidean distances in the RGB color space are not a good approximation of the distances between colors as seen by the
human eye. Instead the L*a*b* color space seems to do a better job (the euclidean distances between L*a*b* colors
are a lot more similar to the distances between colors as perceived by the human eye). L* encodes lightness,
a* red to green and b* blue to yellow. Therefore we used as distance the euclidean distance between the coordinates
of the pixels converted from the RGB color space to the L*a*b* color space, which we didn't need to compute it ourselves
as it's already in skimage (function deltaE_cie76 in skimage.color).

In order to define the threshold distance to tell apart coat and not we'll now study the distances with this metrics.
In the following image we display the distances histogram from each pixel to pure black in logarithmic scale,
along with the aribitrarly chosen threshold (75), which looks quite natural.
'''

from skimage.color import deltaE_cie76
import seaborn as sns

distances = deltaE_cie76(coat, np.zeros_like(coat))

threshold = 75

plt.figure(figsize=(20, 5))

plt.yscale('log')
plt.xlabel('Distance to black')
plt.title('Logarithmic yscale histogram of distances to black frequencies')
plt.xticks(range(0, 301, 25))

sns.histplot(distances.flatten(), bins=50)

plt.axvline(threshold, color='red', label='Empiric threshold')
plt.text(63.5, 1000, 'Empiric threshold')

plt.show()
```

```python
'''
Let's now analyze the difference with this slighly complex distance splitting vs just choosing pure black.
In the following image we see the pixels (augmented in order to see them) which this metrics chooses as
not coat that would be considered coat with the pure black metric. They are all part of the edge of the
coat. They were likely really black to the human eye and would be a bit weird in the yellow coat, but
it doesn't make a whole lot of difference
'''

from scipy.signal import convolve2d

low_distances = distances * (distances > 0) * (distances < threshold)

plt.imshow(convolve2d(low_distances, np.ones((10, 10))), cmap='gray')
plt.title('Pixels exluded from the pure black approach')
plt.axis('off')
plt.show()
```

```python
# Finally, let's define the mask with the chosen threshold

from skimage.transform import rescale
from skimage import img_as_ubyte

coat_mask = (distances > threshold).reshape((distances.shape[0], distances.shape[1], 1))

# We take the pixels where coat_mask==True from the image coat and the others from the model image
coat_model = coat * coat_mask + model * ~coat_mask

# The other pattern, green leaves. We need to make some adjustements in order to look good
# (make the leaves smaller). We do it by scaling down the image.
hawaiian_pattern_original = io.imread('images_notebook/hawaiian_pattern.jpg')
hawaiian_pattern = img_as_ubyte(rescale(hawaiian_pattern_original, 0.6, channel_axis=2))
weed_model = hawaiian_pattern[:model.shape[0], :model.shape[1]] * coat_mask + model[:, :, 0:3] * ~coat_mask

fig, axs = plt.subplots(ncols=2, figsize=(16, 12))

axs[0].imshow(coat_model)
axs[0].axis('off')
axs[0].title.set_text('Yellow coat')

axs[1].imshow(weed_model)
axs[1].axis('off')
axs[1].title.set_text('Hawaiian pattern')

plt.show()
```
