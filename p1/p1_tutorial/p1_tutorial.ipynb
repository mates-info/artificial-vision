# Short  Numpy and Skimage Tutorial

This is a very short introduction to [numpy](http://www.numpy.org) and [Scikit-image library](http://scikit-image.org/), including the main operations for image manipulation.

Numpy is the most important scientific package in the Python ecosystem, because it provides a common datastructure on which many other packages are build on.

What is an image? Images are n-dimensional arrays (`ndarray`) where each element corresponds to a pixel. 

The most common image kinds are:

1. **Binary images**, containing 1 channel of X rows and Y columns. They use logical (True/False, 0/1...) values.

2. **Grey-level images**, containing 1 channel of X rows and Y columns. They use a set of values, depending on the data type (float, uint8, uint16...)

3. **RGB /color images**, containing 3 channels per color (RGB). Sometimes, **RGBA (4 channels)** images are used.

Images manipulated by the Scikit-image library are stored as **Numpy arrays**.

 See [A crash course on NumPy for images](http://scikit-image.org/docs/dev/user_guide/numpy_images.html#numpy-indexing) for more information.

### Basic image manipulation

NumPy arrays representing images can be of different numerical types, such as integer or float. 

See [Image data types and what they mean](http://scikit-image.org/docs/dev/user_guide/data_types.html#data-types) for more information about these types and how scikit-image treats them.

How shall we change the type of an image?

Create an image of type **uint8** specifying its pixel values.

```python
ii=np.array([[0, 255, 0],
             [255,0,255],
             [0, 255, 0]],
             dtype=np.uint8)
print(ii)
```

```python
import matplotlib.pyplot as plt

plt.imshow(ii, cmap='gray')
```

```python
print(ii.dtype)
print(ii.shape)
```


Convert the previous image from **uint8** to **float**

```python
ii2 = ii.astype(float)
print(ii2)
```

```python
from skimage import img_as_float

ii_float = img_as_float(ii)
print(ii_float)
```

**What is going on?!** 

What is the difference between the two operations?

```python
print(ii.dtype)
print(ii2.dtype)
print(ii_float.dtype)
```

## Creating a color image

```python
ii=np.zeros((20,30,3), ## 3 channels (RGB)
            dtype=np.uint8)

plt.imshow(ii)
plt.axis('off')
```

We are going to use the function *np.arrange()*

```python
ex = np.arange(0, 100, 10, dtype=np.uint8) # arange function defines a vector from 0 to 50 in steps of 10.
print(ex)
```

We are going to modify the pixels withing the original black image

```python
ii[ 5, 0:29, 0] = 8*np.arange(29)  ## Row 5, columns between 0:29, Channel 0 (=Red)
ii[10, 0:29, 1] = np.arange(29)    ## Row 10, columns between 0:29, Channel 1 (=Green)
ii[15, 0:29, 2] = 8*np.arange(29)+15  ## Row 15, columns between 0:29, Channel 0 (=Blue)

ii[18, 0:29, :] = [255,255,255]    ## Row 18, columns between 0:29, all channels

plt.imshow(ii)
plt.axis('off')
```

## Spatial and photometric resolution: 
### Function `rescale` 

Scikit-image provides a set of [standard test images](http://scikit-image.org/docs/dev/api/skimage.data.html?highlight=imread#skimage.data.imread) in the Module `data`: astronaut, binary_blobs, camera, checkerboard, chelsea, clock, coffee, coins, etc.

Consider the following example:

```python
from skimage import data

coins = data.coins()
```

```python
print(type(coins))
print(coins.dtype)
print(coins.shape) # check number of channels
print(coins.size)
```

```python
plt.imshow(coins)
plt.axis('off')
plt.show()

plt.imshow(coins, cmap='gray') # we can visualize the image 'camera' in gray scale colormap, 
                                #you can look at matplotlib for different colormaps
plt.axis('off')
plt.show()
```

Sometimes we need to change the spatial and photometric resolution of an image. For this, we will use functions like rescale and resize.

### Rescale function

```python
from skimage.transform import rescale

rescaled = rescale(coins, 0.5)
coins_float=img_as_float(coins)

print('dtypes: ', coins.dtype, coins_float.dtype, rescaled.dtype)
print('min: ', coins.min(), coins_float.min(), np.round(rescaled.min(),4))
print('max: ', coins.max(), coins_float.max(), np.round(rescaled.max(),4))
print('shape: ', coins.shape, coins_float.shape, rescaled.shape)
```

### How to preserve the grayscale range?

```python
rescaled2 = rescale(coins, 0.5, preserve_range=True)

print('dtypes: ', coins.dtype,  rescaled2.dtype)
print('min: ', coins.min(),  np.round(rescaled2.min(),4))
print('max: ', coins.max(), np.round(rescaled2.max(),4))
print('shape: ', coins.shape,  rescaled2.shape)
```

### Resize function

```python
from skimage.transform import resize
from skimage import img_as_ubyte # img_as_ubyte converts an image to 8-bit unsigned integer format

resized=resize(coins,(48,48))
resized_int = img_as_ubyte(resized) 

plt.imshow(resized_int, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.show()
```

```python
print('dtypes: ', coins.dtype, coins_float.dtype, resized_int.dtype)
print('min: ', coins.min(), resized.min(), np.round(resized_int.min(),4))
print('max: ', coins.max(), resized.max(), np.round(resized_int.max(),4))
print('shape: ', coins.shape, resized.shape, resized_int.shape)
```

## Image conversion: 

### From RGB to Grayscale

Often we will need rto convert color images to gray level images. Converting an RGB image to a grayscale image is per with `rgb2gray()`

```python
from skimage.color import rgb2gray

img = data.astronaut()
plt.imshow(img)
plt.axis('off')
plt.show()

print ("Note that dtype: ", img.dtype, "pixel.min: ", img.min(), "pixel.max: ", img.max()) 
```

```python
img_gray = rgb2gray(img)

plt.imshow(img_gray, cmap='gray')
plt.axis('off')
plt.show()

print ("dtype: ", img_gray.dtype, "gray.min: ", img_gray.min(), "gray.max: ", img_gray.max()) 
```

**Question:** Given a gray image, how would it be looking if it is converted to RGB values? 

###Conversion from RGBA to RGB 

Often you can find color images in format [RGBA](https://en.wikipedia.org/wiki/Alpha_compositing#Alpha_blending). **RGBA** has 24 bits for **RGB** color, bit there is an additional 8 bit value for transparency.

The forth component shows the coefficient of blending of each channel. Converting an **RGBA** image to an **RGB** image by alpha blending it with a background is performend by the command `rgba2rgb()`.

```python
img_rgba = data.logo()

print("dtype: ", img_rgba.dtype)
print('shape: ', img_rgba.shape) ## check the number of channels

plt.imshow(img_rgba)
plt.axis('off')
plt.show()
```

```python
from skimage.color import rgba2rgb

img_rgb = rgba2rgb(img_rgba)

print("dtype: ", img_rgb.dtype)
print('shape: ', img_rgb.shape) ## check the number of channels

plt.imshow(img_rgb)
plt.axis('off')
plt.show()
```
