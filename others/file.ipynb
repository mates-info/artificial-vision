```python
from sklearn.cluster import KMeans
import numpy as np

rocki_5channels = np.zeros((rocki.shape[0], rocki.shape[1], 5))
rocki_5channels[:, :, :3] = rocki
rocki_5channels[:, :, 3] = np.arange(rocki.shape[0]).reshape(rocki.shape[0], 1) / rocki.shape[0] * 100
rocki_5channels[:, :, 4] = np.arange(rocki.shape[1]).reshape(1, rocki.shape[1]) / rocki.shape[1] * 100

rocki_1d = rocki_5channels.reshape(rocki.shape[0] * rocki.shape[1], 5)

kmeans = KMeans(n_clusters=3).fit(rocki_1d)

classes = kmeans.labels_.reshape((rocki.shape[0], rocki.shape[1]))

io.imshow(classes)
plt.show()
```
